```python
df_x = diabetes.data
df_y = diabetes.target
```

```python
import numpy as np
df_x = np.array(df_x)
df_y = np.array(df_y)
```

```python
from sklearn.model_selection import train_test_split

df_x_train, df_x_test, df_y_train, df_y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
```

```python
w = np.random.randn(10)
b = np.random.randn()
w
```

```python
class DiabetesMLR:
    def __init__(self, learning_rate=0.1):
        self.w = None  # 모델의 가중치 (weight) 벡터
        self.b = None  # 모델의 바이어스 (bias)
        self.lr = learning_rate  # 학습률 (learning rate)
        self.losses = []  # 매 에포크마다 손실을 저장하기 위한 리스트
        self.w_hist = []  # 매 에포크마다 계산된 가중치(weight)를 저장하기 위한 리스트
        self.b_hist = []  # 매 에포크마다 계산된 바이어스(bias)를 저장하기 위한 리스트
    
    def forward(self, df_x):
        # 모델 예측: 선형 회귀 식 y_pred = X * weights + bias
        y_pred = np.dot(df_x, self.w) + self.b
        return y_pred

    def loss(self, df_x, df_y):
        # 손실 계산: 평균 제곱 오차(MSE)
        y_pred = self.forward(df_x)
        return np.mean((y_pred - df_y) ** 2)

    def gradients(self, df_x, df_y):
        # 가중치 및 바이어스에 대한 그래디언트(기울기) 계산
        y_pred = self.forward(df_x)
        error = y_pred - df_y
        weight_grad = 2 * np.dot(df_x.T, error)
        bias_grad = 2 * np.sum(error)
        return weight_grad, bias_grad

    def fit(self, df_x_train, df_y_train, epochs=20):
        # 가중치와 바이어스를 초기화
        self.w = np.ones(df_x_train.shape[1])  # 모든 가중치를 1로 초기화
        self.b = 0  # 바이어스를 0으로 초기화

        for epoch in range(epochs):
            # 손실 및 그래디언트 초기화
            total_loss = 0
            weight_grad = np.zeros(df_x_train.shape[1])
            bias_grad = 0

            # 각 샘플에 대해 손실 및 그래디언트 계산
            for df_x, df_y in zip(df_x_train, df_y_train):
                total_loss += self.loss(df_x, df_y)
                w_grad, b_grad = self.gradients(df_x, df_y)
                weight_grad += w_grad
                bias_grad += b_grad

            # 가중치 및 바이어스 업데이트
            self.w -= self.lr * (weight_grad / len(df_y_train))
            self.b -= self.lr * (bias_grad / len(df_y_train))

            # 손실 및 매개변수 저장
            avg_loss = total_loss / len(df_y_train)
            print(f'Epoch {epoch + 1}/{epochs} ===> Loss: {avg_loss:.5f}')
            self.losses.append(avg_loss)
            self.w_hist.append(self.w.copy())
            self.b_hist.append(self.b)

        
    
```

```python
model = DiabetesMLR(learning_rate=0.01)
model.fit(df_x, df_y, epochs=30000)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 테스트 데이터에 대한 성능 확인
def evaluate_model(model, df_x_test, df_y_test):
    # 모델을 사용하여 테스트 데이터에 대한 예측 수행
    y_pred = model.forward(df_x_test)
    
    # 손실 계산 (평균 제곱 오차, MSE)
    mse = np.mean((y_pred - df_y_test) ** 2)
    print(f'Test MSE: {mse:.5f}')
    
    return y_pred

# 예시 테스트 데이터 (df_x_test, df_y_test)
# 실제 데이터를 사용해 평가할 경우, df_x_test와 df_y_test를 할당해주세요.

y_test_pred = evaluate_model(model, df_x_test, df_y_test)
```

```python
def plot_predictions(y_true, y_pred, title='Model Predictions vs True Values'):
    plt.figure(figsize=(10, 6))
    plt.plot(y_true, label='True Values', marker='o')
    plt.plot(y_pred, label='Predicted Values', marker='x')
    plt.title(title)
    plt.xlabel('Samples')
    plt.ylabel('Values')
    plt.legend()
    plt.show()

# 정답 데이터와 예측 데이터 시각화
plot_predictions(df_y_test, y_test_pred)
```
